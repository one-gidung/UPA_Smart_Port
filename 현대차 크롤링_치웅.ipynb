```python
chrome_path = chrome_path
driver = webdriver.Chrome(chrome_path)

url = 'https://www.hyundai.com/kr/ko/e/vehicles/avante/intro'
driver.get(url)
time.sleep(2)

driver.maximize_window()
```

```python
file_no = 0                                
count = 1
img_src2=[]
img_name = []

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
img_src = soup.find('div','vr_car_preload_image_container').find_all('img') ##대부분 img태그에 들어있음.
##사진 url주소 찾는게 젤 중요함.
```

```python
#이미지 주소따오기
for i in img_src :
        img_src1=i['data-src']
        img_src2.append(img_src1)
        count += 1
```

```python
#이미지 이름
for i in img_src :
        img_src1=i['alt']
        img_name.append(img_src1)
        count += 1
```

```python
for i in range(0,len(img_src2)) :
        try :
                urllib.request.urlretrieve('https://www.hyundai.com/'+img_src2[i+1], str(img_name)+str(file_no)+'.png') ##(주소, 경로&이름)
        except :
                continue  #썸네일만 가져오면 상관없는데, 원본에 있는 주소일때, 없는경우가 있음. 그래서 예외처리가 필수임      
        file_no += 1                
        time.sleep(0.2)      
        print("%s 번째 이미지 저장중입니다=======" %file_no)
```
