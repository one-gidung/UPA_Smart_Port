# 데이터 EDA를 하기 위한 전처리

```python
# 날짜데이터들 timestamp(datetime)형으로 변환
for colname in ['입항일시', '출항일시', 'CIO수속일자', '수리일시']:
    df[colname] = pd.to_datetime(df[colname], format='%Y-%m-%d %H:%M')
df['입항일시']
```

```python
# 입항횟수(연도) timestamp(datetime)형으로 변환
df['입항횟수'] = pd.to_datetime(df['입항횟수'], format='%Y')
```

```python
# 톤수들의 문자열에서 ',' 제거 후 numeric으로 변환
for colname in ['총톤수', '국제톤수', '징수톤수']:
    df[colname] = df[colname].str.replace(',', '')
    df[colname] = pd.to_numeric(df[colname])
```

```python
# 필요없는 변수 제거
df.drop(['항명', '입항횟수', '구분', '외내', '항해구분', '예선', '도선', '국적', '국적.1', '승무원 1', '승무원 2', '승객', '부선호출부호1', '부선호출부호2', '계선장소.2'], axis=1, inplace=True)
df
```

```python
# 입출 칼럼 입항:1, 출항:0 으로 변환

# dict: 기존 값, 바꾸고자 하는 값
change_dict = {'입항': 1, '출항': 0}

# 변경
df = df.replace({'입출':change_dict})
df
```

```python
# 칼럼 이름 변경
df.rename(columns={'입항횟수.1': '입항횟수'}, inplace=True)
df
```

```python
# 계선장소.1 string으로 변경후 계선장소와 합침
# 계선장소.1 object로 타입변환
df['계선장소.1'] = df['계선장소.1'].astype(str)
# 새로운 열 생성
df['계선장소.2'] = df['계선장소.1'].str[0]
df['계선장소.1'] = "0" + df['계선장소.2'].map(str)
```

```python
# 변환하기 위해 만들었던 열 삭제
df.drop('계선장소.2', axis=1, inplace=True)
df
```

```python
df['계선'] = df['계선장소'] + " " +  df['계선장소.1'].astype(str)
df
```

```python
df.drop(['계선장소.1'], axis=1, inplace=True)
df
```
