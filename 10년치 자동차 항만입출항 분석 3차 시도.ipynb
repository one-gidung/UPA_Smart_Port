# 데이터 EDA를 하기 위한 전처리

```python
# 필요없는 변수 제거
df.drop(['항명', '입항횟수', '구분', '외내', '항해구분', '예선', '도선', '국적', '국적.1', '승무원 1', '승무원 2', '승객', '부선호출부호1', '부선호출부호2', '계선장소.2', 'Unnamed: 4', 'Unnamed: 18', 'Unnamed: 20', 'Unnamed: 21'], axis=1, inplace=True)
df
```

```python
# 선박용도가 자동차운반선인것들의 데이터 프레임 = df_Car
df_Car = df.query('선박용도 == "자동차운반선"').reset_index(drop=True)
df_Car
```

```python
# 날짜데이터들 timestamp(datetime)형으로 변환
for colname in ['입항일시', '출항일시', 'CIO수속일자', '수리일시']:
    df_Car[colname] = pd.to_datetime(df_Car[colname], format='%Y-%m-%d %H:%M:%S')
df_Car['입항일시']
```

```python
# 1950년 1월1일 입항인 선박 행 삭제
df_Car.drop(df_Car.query('입항일시 == "1950-01-01 00:00:00"').index, inplace=True)
df_Car
```

```python
# 톤수들의 문자열에서 ',' 제거 후 numeric으로 변환
for colname in ['총톤수', '국제톤수', '징수톤수']:
    df_Car[colname] = df_Car[colname].str.replace(',', '')
    df_Car[colname] = pd.to_numeric(df_Car[colname])
```

```python
# 입출 칼럼 입항:1, 출항:0 으로 변환

# dict: 기존 값, 바꾸고자 하는 값
change_dict = {'입항': 1, '출항': 0}

# 변경
df_Car = df_Car.replace({'입출':change_dict})
df_Car
```

```python
# 칼럼 이름 변경
df_Car.rename(columns={'입항횟수.1': '입항횟수'}, inplace=True)
df_Car
```

```python
# 계선장소.1 string으로 변경후 계선장소와 합침
# 계선장소.1 object로 타입변환
df_Car['계선장소.1'] = df_Car['계선장소.1'].astype(str)
# 새로운 열 생성
df_Car['계선장소.2'] = df_Car['계선장소.1'].str[0]
df_Car['계선장소.1'] = "0" + df_Car['계선장소.2'].map(str)
```

```python
# 변환하기 위해 만들었던 열 삭제
df_Car.drop('계선장소.2', axis=1, inplace=True)
df_Car
```

```python
# 새로운 열 추가
df_Car['계선'] = df_Car['계선장소'] + " " +  df_Car['계선장소.1'].astype(str)
df_Car
```

```python
# 필요없는 열 삭제 후 인덱스 초기화
df_Car.drop(['계선장소.1'], axis=1, inplace=True)
df_Car.reset_index(inplace=True, drop=True)
df_Car
```

```python
# 입출항 차이 행 추가
df_Car['입출항 차이'] = df_Car['출항일시']-df_Car['입항일시']
df_Car
```
