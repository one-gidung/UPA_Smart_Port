```python
# 필요없는 변수 제거
df_car.drop(['항명', '구분', '외내', '국제톤수', '징수톤수', 'CIO수속일자', '수리일시', '선박용도', '항해구분', '국적', '국적.1', '승무원 1', '승무원 2', '승객', '부선호출부호1', '부선호출부호2', '계선장소.2'], axis=1, inplace=True)
df_car
```

```python
# 입출 칼럼 입항:1, 출항:0 으로 변환

# dict: 기존 값, 바꾸고자 하는 값
change_dict = {'입항': 1, '출항': 0}

# 변경
df_car = df_car.replace({'입출':change_dict})
df_car
```

```python
# 칼럼 이름 변경
df_car.rename(columns={'입항횟수.1': '입항횟수', '입항횟수':'입항년도'}, inplace=True)
df_car
```

```python
df_car.info()
```

```python
df_car.isnull().sum()
```

```python
# MRN에서 선사번호 추출, 새로운 열 생성
df_car['선사번호'] = df_car['MRN 번호'].str[2:6]

# # MRN 번호 드랍
df_car.drop('MRN 번호', axis=1, inplace=True)

# 예선도선 칼럼 Y:1, N:0 으로 변환
# dict: 기존 값, 바꾸고자 하는 값
change_dict = {'Y': 1, 'N': 0}

# 변경
df_car = df_car.replace({'예선': change_dict,
                '도선': change_dict})

# int형으로 변환하기 위해 결측치 제거(2개 밖에 없으므로 그냥 drop)
df_car.dropna(subset=['예선', '도선'], inplace=True)

# 예선, 도선 int형으로 변환
df_car[['예선', '도선']] = df_car[['예선', '도선']].astype(int)

# index 초기화 시켜주기
df_car = df_car.reset_index(drop=True)
df_car
```

```python
# 날짜데이터들 timestamp(datetime)형으로 변환
for colname in ['입항일시', '출항일시']:
    df_car[colname] = pd.to_datetime(df_car[colname], format='%Y-%m-%d %H:%M:%S')

# 1950년 1월1일 입항인 선박 행 삭제
df_car.drop(df_car.query('입항일시 == "1950-01-01 00:00:00"').index, inplace=True)
```

```python
# 입출항 차이 행 추가
df_car['작업시간'] = df_car['출항일시']-df_car['입항일시']

# 입출항 차이가 0보다 작은 행 제외
ind = df_car[df_car['작업시간'] <= '0 days 00:00:00'].index
df_car.drop(ind, inplace=True)

# 작업시간 자료형 변환
time = []
for i in range(len(df_car['작업시간'])):
    time.append(float(df_car['작업시간'].iloc[i].days * 24) + float(df_car['작업시간'].iloc[i].seconds / 3600))
df_car['작업시간'] = time

df_car
```

```python
# 입항일시의 월을 추출에 월 컬럼 생성
df_car['입항월'] = df_car['입항일시'].astype(str).str[5:7].astype(int)

df_car = df_car.reset_index(drop=True)
df_car
```
