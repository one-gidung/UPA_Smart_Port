# 데이터 전처리

```python
import pandas as pd

data2012 =pd.read_excel('Port_2012.xlsx',skiprows=12)
data2013 =pd.read_excel('Port_2013.xlsx',skiprows=12)
data2013_2 =pd.read_excel('Port_2013(2).xlsx',skiprows=12)
data2014 =pd.read_excel('Port_2014.xlsx',skiprows=12)
data2015 =pd.read_excel('Port_2015.xlsx',skiprows=12)
data2015_2 =pd.read_excel('Port_2015(1).xlsx',skiprows=12)
data2016 =pd.read_excel('Port_2016.xlsx',skiprows=12)
data2017 =pd.read_excel('Port_2017.xlsx',skiprows=12)
data2018 =pd.read_excel('Port_2018.xlsx',skiprows=12)
data2019 =pd.read_excel('Port_2019.xlsx',skiprows=12)
data2020 =pd.read_excel('Port_2020.xlsx',skiprows=12)
data2021 =pd.read_excel('Port_2021.xlsx',skiprows=12)

li = [data2012,data2013,data2013_2,data2014,data2015,data2015_2,data2016,data2017,data2018,data2019,data2020,data2021]
```

```python
data = pd.concat(li, axis=0, ignore_index=True)
```

```python
data_car = data[data['선박용도'] == '자동차운반선']
```

```python
# data_car.to_excel(excel_writer='C:/Users/ktaehyun/yes1.xlsx')
```

```python
data_car
```

```python
data_car.columns
```

```python
data_car.drop(['항명','입항횟수','구분','외내','국제톤수','징수톤수','항해구분','국적.1','계선장소.2','선박용도',
               '승무원 1','예선','도선','승무원 2','승객','부선호출부호1','부선호출부호2','Unnamed: 4',
               'Unnamed: 18','Unnamed: 20','Unnamed: 21'], axis=1, inplace=True)
data_car.rename(columns={'입항횟수.1':'입항횟수', '계선장소.1':'계선장소번호'}, inplace=True)
data_car.head()
```

```python
data_car['계선장소번호'] = data_car['계선장소번호'].fillna('0')
```

```python
data_car = data_car.astype({'계선장소번호':'int'})
data_car = data_car.astype({'계선장소번호':'str'})
data_car.info()
```

```python
cols = ['계선장소', '계선장소번호']
data_car['계선장소0'] = data_car[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
data_car.drop(['계선장소','계선장소번호'], axis=1, inplace=True)
data_car.rename(columns={'계선장소0':'계선장소'}, inplace=True)
```

```python
data_car['총톤수0'] = data_car.총톤수.str.replace(',','').astype('int')
data_car.drop(['총톤수'], axis=1, inplace=True)
data_car.rename(columns={'총톤수0':'총톤수'}, inplace=True)
data_car.head()
```

```python
data_car['입항횟수'] = data_car['입항횟수'].fillna(0)
data_car = data_car.astype({'입항횟수':'int'})
data_car.info()
```

```python
data_car['입출'] = pd.get_dummies(data_car['입출'], drop_first=True)
data_car
```

```python
data_car['입항일시'] = pd.to_datetime(data_car['입항일시'])
```

```python
data_car['출항일시'] = pd.to_datetime(data_car['출항일시'])
```

```python
data_car.info()
```

```python
data_car
```

```python
data_car.drop(data_car.index[-17:],inplace=True)
data_car.drop(data_car.index[8902:8905],inplace=True)
data_car.drop(data_car.index[8193],inplace=True)
data_car.drop(data_car.index[0:12],inplace=True)
```

```python
# data_car.to_excel(excel_writer='C:/Users/ktaehyun/yes2.xlsx')
```

```python
data_car
```

```python
data_car['작업시간'] = data_car['출항일시'] - data_car['입항일시']
```

```python
data_car.drop(['입항일시', '출항일시'], axis=1, inplace=True)
data_car
```

```python
import datetime as dt

data_car['작업시간(Minute)'] = data_car['작업시간'].dt.total_seconds()/60
data_car = data_car.astype({'작업시간(Minute)':'int'})
data_car['작업시간(Minute)'] = data_car['작업시간(Minute)'].abs()
```

```python
data_car.drop(data_car.index[7814],inplace=True)
data_car.drop(data_car.index[7785],inplace=True)
data_car.drop(data_car.index[6388],inplace=True)
data_car.drop(data_car.index[6337],inplace=True)
data_car.drop(data_car.index[5951:5953],inplace=True)
data_car.drop(data_car.index[5021:5023],inplace=True)
data_car.drop(data_car.index[4530],inplace=True)
data_car.drop(data_car.index[4236],inplace=True)
```
