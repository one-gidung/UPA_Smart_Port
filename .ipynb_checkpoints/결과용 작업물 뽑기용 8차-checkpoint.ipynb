```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

font_location = 'C:/Windows/Fonts/malgunsl.ttf'
font_name = fm.FontProperties(fname = font_location).get_name()
sns.set(rc = {'figure.figsize':(15,8)})
matplotlib.rc('font' , family=font_name)#한글폰트 설정
```

```python
df = pd.read_excel('data/자동차항만입출항전처리데이터.xlsx', index_col=0)
df
```

```python
A = df[['입항년도', '입항월', '입항횟수']].groupby(['입항년도', '입항월']).count()
A
```

```python
A.drop(2011, inplace=True)
```

```python
A = A.reset_index()
A
# A.to_index()
```

```python

for j in range(9):
    plt.plot(A['입항월'].iloc[:12], A['입항횟수'].iloc[12*j:12*j+12], label=str(2012+j))


plt.plot(A['입항월'].iloc[:7], A['입항횟수'].iloc[-8:-1], label="2021")
plt.legend()
```

```python
A["입항일시"] = A['입항년도'].astype(str) + "-" + A['입항월'].astype(str)
A["입항일시"] = pd.to_datetime(A["입항일시"])
A
```

```python
A[['입항횟수']].groupby(['입항일시']).count()
```

```python
plt.plot(A['입항일시'].iloc[:-1], A['입항횟수'].iloc[:-1], label="입출항횟수")
plt.legend()
```
