# 변수 정의

항명: 선박이 국내에 들어왔을때의 항구의 이름(ex: 울산항)\
호출부호: 무선국을 식별하기 위한, 중복되지 않는 일련의 문자열이다.호출부호는 앞부분의 프리픽스와 뒷부분의 서픽스로 구성된다. 그래서 콜사인(Call Sign)이라고 한다.\
선명: 선박의 이름(한,영 혼용돼있음)\
입항횟수: 입항년도\
입항횟수.1: 입항횟수\
구분: 최초(입출항 24시간 전) / 최종(입출항 확정시) / 변경 / 취소(거의 없음) \
외내: 외항 / 내항 \
입출: 입항 / 출항 \
총톤수: 선박의 크기를 나타내는 지표, 4000톤 기준으로 조금씩?? \
국제톤수: \
징수톤수: 선박을 등록하거나 세금 및 각종 요금의 징수 기준을 삼을 때의 톤수(=법정톤수)\
입항일시\
출항일시\
CIO수속일자: 최고정보책임자(Chief Information Officer)\
수리일시: ??\
항해구분: 신조선 / 기타\
MRN 번호: 적하목록관리번호(Manifest Reference Number), 화물관리번호(제출년도 2자리 숫자 + 운항선사 영문코드 4자리 + 선사에서 부여한 고유번호 4자리 + 오류검증 번호 1자리) 총 11자리\
국적: 선박의 국적(영문이니셜2자리)\
국적.1: 선박의 국적(한글표기)\
계선장소: 부두코드(계류시설:M, 선석:B, 자동차부두:J, 6부두:6) "계선"(계선)이란 선박이 운항을 중지하고 정박하거나 계류하는 것을 말한다.\
계선장소.1: 몇번째 선석인지(개별 자산)\
계선장소.2: 부두 이름+선석 순서\
차항지: 선박이 현재 항구를 떠나 가려고 하는 다음 목적지? or 최종 도착지?(지역명)\
전출항지: 선박이 현재 항구를 오기 전에 있었던 항구(항구명?도시명?)\
선박용도\
승무원 1: 한국인\
승무원 2: 외국인\
승객\
예선: 선박자체의 동력으로 이안, 접안이 곤란한 대형화물 및 여객선과 위험물운송선박 등을 항만시설 보호와 선박안전을 위해 이안, 접안 하거나 입출항할 수 있도록 대상 선박을 밀거나 끌어주는 고마력의 엔진을 가진 특수선박 \
도선: 선장이 모든 항구의 주변 환경을 다 알 수 없으므로 항구 상황을 자세히 알고 선박을 안전한 뱃길로 인도하는 선박 또는 그런 행위, 필수로 도선사용을 해야하나 면제 신청을 하는 경우도 있다. \
"도선"이란 도선구에서 도선사가 선박에 승선하여 그 선박을 안전한 수로로 안내하는 것을 말한다. \
부선호출부호1: \
부선호출부호2: \
여기서  "부선"이라 함은 추진장치가 설치되지 아니하여 자력 항해능력이 없어 다른 선박에 의하여 끌리거나 밀려서 항해되는 선박을 말한다.

```python
# 톤수들의 문자열에서 ',' 제거
for col in ['총톤수', '국제톤수', '징수톤수']:
    df_J[col] = df_J[col].str.replace(',', '')
df_J
```

```python
# object변수인 톤수 컬럼들을 int로 변환
for col in ['총톤수', '국제톤수', '징수톤수']:
    df_J[col] = pd.to_numeric(df_J[col])
df_J.info()
```
