# 데이터 EDA를 하기 위한 전처리

```python
# 선박용도가 자동차운반선인것들의 데이터 프레임 = df_Car
df_Car = df.query('선박용도 == "자동차운반선"').reset_index(drop=True)
df_Car
```

```python
# 필요없는 변수 제거
df_Car.drop(['항명', '구분', '외내', '항해구분', '국적', '국적.1', '승무원 1', '승무원 2', '승객', '부선호출부호1', '부선호출부호2', '계선장소.2'], axis=1, inplace=True)
df_Car
```

```python
# 날짜데이터들 timestamp(datetime)형으로 변환
for colname in ['입항일시', '출항일시', 'CIO수속일자', '수리일시']:
    df_Car[colname] = pd.to_datetime(df_Car[colname], format='%Y-%m-%d %H:%M:%S')
df_Car['입항일시']
```

```python
# 1950년 1월1일 입항인 선박 행 삭제
df_Car.drop(df_Car.query('입항일시 == "1950-01-01 00:00:00"').index, inplace=True)
df_Car
```

```python
# 톤수들의 문자열에서 ',' 제거 후 numeric으로 변환
for colname in ['총톤수', '국제톤수', '징수톤수']:
    df_Car[colname] = df_Car[colname].str.replace(',', '')
    df_Car[colname] = pd.to_numeric(df_Car[colname])
```

```python
# 입출 칼럼 입항:1, 출항:0 으로 변환

# dict: 기존 값, 바꾸고자 하는 값
change_dict = {'입항': 1, '출항': 0}

# 변경
df_Car = df_Car.replace({'입출':change_dict})
df_Car
```

```python
# 칼럼 이름 변경
df_Car.rename(columns={'입항횟수.1': '입항횟수', '입항횟수':'입항년도'}, inplace=True)
df_Car
```

```python
# 계선장소.1 string으로 변경후 계선장소와 합침
# 계선장소.1 object로 타입변환
df_Car['계선장소.1'] = df_Car['계선장소.1'].astype(str)
# 새로운 열 생성
df_Car['계선장소.2'] = df_Car['계선장소.1'].str[0]
df_Car['계선장소.1'] = "0" + df_Car['계선장소.2'].map(str)
```

```python
# 변환하기 위해 만들었던 열 삭제
df_Car.drop('계선장소.2', axis=1, inplace=True)
df_Car
```

```python
# 새로운 열 추가
df_Car['계선'] = df_Car['계선장소'] + " " +  df_Car['계선장소.1'].astype(str)
df_Car
```

```python
# 입출항 차이 행 추가
df_Car['입출항 차이'] = df_Car['출항일시']-df_Car['입항일시']
df_Car
```

```python
# 필요없어보이는 변수 제거
df.drop(['수리일시', '선박용도', '징수톤수', '국제톤수'], axis=1, inplace=True)
df
```

```python
# 입출항 차이가 0보다 작은 행 제외
ind = df[df['입출항 차이'] <= '0 days 00:00:00'].index
df.drop(ind, inplace=True)
df.reset_index(drop=True)
```

```python
# 입출항 차이
time = []
for i in range(len(df['입출항 차이'])):
    time.append(float(df['입출항 차이'].iloc[i].days * 24) + float(df['입출항 차이'].iloc[i].seconds / 3600))

df['입출항 차 (시간)'] = time

df.drop(['입출항 차이'], axis=1, inplace=True)
df.reset_index(drop=True)
df
```

```python
# 입항일시의 월을 추출에 월 컬럼 생성
df['입항월'] = df['입항일시'].astype(str).str[5:7].astype(int)

df = df.reset_index(drop=True)
df
```

```python
# 컬럼 순서 바꾸기 및 필요없는 열 제거
df = df[['선명', '호출부호', '입항년도', '입항월', '입항횟수', '입출', '총톤수', '계선', '계선장소', '계선장소.1', '입출항 차 (시간)', '예선', '도선', '차항지', '전출항지', 'MRN 번호']]
df = df.reset_index(drop=True)
df
```

```python
# 칼럼 이름 변경
df.rename(columns={'입출항 차 (시간)': '작업시간'}, inplace=True)
df
```

```python
# MRN에서 선사번호 추출, 새로운 열 생성
df['선사번호'] = df['MRN 번호'].str[2:6]

# MRN 번호 드랍
df.drop('MRN 번호', axis=1, inplace=True)
df = df.reset_index(drop=True)
df
```
